```python
from diffusers import StableDiffusionPipeline
import torch
```

```python
model_id = "models/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(model_id)
pipe = pipe.to("cuda")
```

```python
prompt = "Donald Trump eating a Krabby Patty with Squidward"
model_output = pipe(prompt)
```

```python
model_output.images[0].save('data/beans shapiro.png')
```
